```python
df = pd.read_csv("/content/drive/MyDrive/data/DataCoSupplyChainDataset.csv",encoding= 'unicode_escape')
df.head()
```

```python
df
```

```python
df = df.drop(['Customer Email','Product Image','Customer Zipcode','Product Status','Order Zipcode','Customer Password'],axis=1)
```

```python
df = df.drop(['Product Description'],axis=1)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
string_data = list(df.select_dtypes(include=['object']).columns)
print(string_data)
df[string_data]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[string_data] = df[string_data].apply(le.fit_transform)
```

```python
df.fillna(0)
```

```python
X = df.drop('Late_delivery_risk',axis=1)
y = df['Late_delivery_risk']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn import preprocessing
scaler_x = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler_x.transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0,max_iter=1000).fit(X_scaled,y_train)
```

```python
scaler_x = preprocessing.StandardScaler().fit(X_test)
X_scaled_test = scaler_x.transform(X_test)
```

```python
clf.score(X_scaled_test,y_test)
```
